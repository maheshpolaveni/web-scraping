```python
import requests
from bs4 import BeautifulSoup
url='https://en.wikipedia.org/wiki/2019_Cricket_World_Cup'
response=requests.get(url)    
content=response.text
soup=BeautifulSoup(content,'lxml')
# finding the div elemnts with the specific inline style
div_elements = soup.find_all('div', {'style': 'width: 100%; background-color: #eee; clear:both'})
#extracting the text content of all div_elements within each div element
for div in div_elements:
    text=div.text
    print(text)
    
```
